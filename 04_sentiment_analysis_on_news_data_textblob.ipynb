```python
import numpy as np
import pandas as pd
import pickle
```

```python
from textblob import TextBlob
```

```python
import matplotlib.pyplot as plt
# %matplotlib
```

### Import the previous data

```python
news_data = pd.read_pickle('pickle/clean_news_data.pkl')
type(news_data), len(news_data)
```

## Sentiment Analysis

### Function to Analyse Sentiment

```python
def analyze_sentiment(text):
    text = TextBlob(text)
    return text.sentiment
```

### Title

```python
title_polarity = list()
title_subjectivity = list()
for i in range(50):
    data = analyze_sentiment(news_data['title'][i])
    title_polarity.append(data[0])
    title_subjectivity.append(data[1])
print('Polarity:', title_polarity, '\n')
print('Subjectivity:', title_subjectivity, '\n')
```

```python
content_polarity = list()
content_subjectivity = list()
for i in range(50):
    data = analyze_sentiment(news_data['content'][i])
    content_polarity.append(data[0])
    content_subjectivity.append(data[1])
print('Polarity:', content_polarity, '\n')
print('Subjectivity:', content_subjectivity, '\n')
```

---

## Graph Representation
---

### Graph between Polarity of Title and Content
This graph describe the polarity of the news title and content. <br>
<br>
The higher the graph is the more it is likely to be positive. <br>
The lower the graph is the more it is likely to be negative.

```python
plt.rcParams['figure.figsize'] = 20, 5
x = np.arange(50)
width = 0.35
fig, ax = plt.subplots()

rec1 = ax.bar(x - width/2, title_polarity, width, label='Title', color='orange')
rec2 = ax.bar(x + width/2, content_polarity, width, label='Content', color='blue')

ax.set_ylabel('Polarity')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend()

plt.show()
```

```python
plt.rcParams['figure.figsize'] = 20, 5
plt.plot(title_polarity, 'orange')
plt.plot(content_polarity, 'blue')
```

### Conclusion
The graph depict the polarity of the news title and content. <br>
- Greater then '0' : Positive
- Lesser then '0' : Negative
---

### Graph between Subjectivity of Title and Content
This graph describe the subjectivity of the news title and content.<br>
<br>
The higher the graph is the more it is likely to be opinion-based.<br>
The lower the graph is the more it is likely to be fact-based.<br>

```python
plt.rcParams['figure.figsize'] = 20, 5
x = np.arange(50)
width = 0.35
fig, ax = plt.subplots()

rec1 = ax.bar(x - width/2, title_subjectivity, width, label='Title', color='orange')
rec2 = ax.bar(x + width/2, content_subjectivity, width, label='Content', color='blue')

ax.set_ylabel('Subjectivity')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend()

plt.show()
```

```python
plt.rcParams['figure.figsize'] = 20, 5
plt.plot(title_subjectivity, 'orange')
plt.plot(content_subjectivity, 'blue')
```

### Conclusion
This graph depict the subjectivity of the news title and content.<br>
The higher the graph goes the news become more opinion based then fact based
