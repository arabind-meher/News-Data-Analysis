# 04 - Sentiment Analysis on News Data (Subjectivity)

```python
import numpy as np
import pandas as pd
import pickle
```

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
```

```python
import matplotlib.pyplot as plt
# %matplotlib
```

### Import the previous data

```python
news_data = pd.read_pickle('pickle/clean_news_data.pkl')
len(news_data)
```

### Function to Analyse Sentiment

```python
def analyze_sentiment(text):
    text = TextBlob(text)
    return text.sentiment
```

```python
title_polarity = list()
title_subjectivity = list()
for i in range(50):
    data = analyze_sentiment(news_data['title'][i])
    title_polarity.append(data[0])
    title_subjectivity.append(data[1])
len(title_polarity), len(title_subjectivity)
```

```python
content_polarity = list()
content_subjectivity = list()
for i in range(50):
    data = analyze_sentiment(news_data['content'][i])
    content_polarity.append(data[0])
    content_subjectivity.append(data[1])
len(content_polarity), len(content_subjectivity)
```

## Graph Representation

### Graph between Subjectivity of Title and Content
This graph describe the subjectivity of the news title and content.<br>
<br>
The higher the graph is the more it is likely to be opinion-based.<br>
The lower the graph is the more it is likely to be fact-based.<br>

```python
plt.rcParams['figure.figsize'] = 20, 5
x = np.arange(50)
width = 0.35
fig, ax = plt.subplots()

rec1 = ax.bar(x - width/2, title_subjectivity, width, label='Title', color='yellow')
rec2 = ax.bar(x + width/2, content_subjectivity, width, label='Content', color='blue')

ax.set_ylabel('Subjectivity')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend()

plt.show()
```

```python
plt.rcParams['figure.figsize'] = 20, 5
plt.plot(title_subjectivity, 'yellow')
plt.plot(content_subjectivity, 'blue')
```

This graph depict the subjectivity of the news title and content.<br>
