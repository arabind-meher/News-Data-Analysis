## Sentiment Analysis

### Function to Analyze Sentiment

```python
def analyse_sentiment(text):
    analyse = SentimentIntensityAnalyzer()
    sentiment = analyse.polarity_scores(text)
    return sentiment
```

### Title

```python
# Title
title_negative = list()
title_neutral = list()
title_positive = list()
title_compound = list()
for i in range(50):
    data = analyse_sentiment(news_data['title'][i])
    title_negative.append(data['neg'] * 100)
    title_neutral.append(data['neu'] * 100)
    title_positive.append(data['pos'] * 100)
    title_compound.append(data['compound'])
print('Negativity:', title_negative, '\n')
print('Positivity:', title_positive, '\n')
print('Neutral:', title_neutral, '\n')
print('Compound:', title_compound, '\n')
```

### Content

```python
# Content
content_negative = list()
content_neutral = list()
content_positive = list()
content_compound = list()
for i in range(50):
    data = analyse_sentiment(news_data['content'][i])
    content_negative.append(data['neg'] * 100)
    content_neutral.append(data['neu'] * 100)
    content_positive.append(data['pos'] * 100)
    content_compound.append(data['compound'])
print('Negativity:', content_negative, '\n')
print('Positivity:', content_positive, '\n')
print('Neutral:', content_neutral, '\n')
print('Compound:', content_compound, '\n')
```

## Graph Representaton
---

### 'Content Compound' Graph
Bar graph and Line graph both express the overall negativity and positivity of the sentences present Content.<br>
- If value is greater then '0': Positive
- If value is smaller then '0': Negative
- If value is equal to '0': Neutral

```python
plt.rcParams['figure.figsize'] = 20, 5
x = np.arange(50)
width = 0.35
fig, ax = plt.subplots()

bar = ax.bar(x + width/2, content_compound, width, label='Content', color='blue')

ax.set_ylabel('Sentiment')
ax.set_xticks(x)
ax.set_xticklabels(x)
ax.legend()

plt.show()
```

```python
plt.rcParams['figure.figsize'] = 20, 5
plt.plot(content_compound, 'blue')
plt.ylabel('Sentiment')
plt.xlabel('Data')
plt.show()
```

### Conclusion
The above two graphs conclude that the content of the news is either positive or negative. In very few cases the news tends towards neutral.
